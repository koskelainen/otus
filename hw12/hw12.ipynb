```python
import pandas as pd

MIN_YEAR = 1880
MAX_YEAR = 2011
URL_PATTERN = "https://raw.githubusercontent.com/wesm/pydata-book/2nd-edition/datasets/babynames/yob{}.txt"
COLUMN_NAMES = ['name', 'sex', 'num']


def load_src():
    data_set = None
    for year in range(MIN_YEAR, MAX_YEAR):
        if data_set is None:
            data_set = pd.read_csv((URL_PATTERN.format(year)), header=None, names=COLUMN_NAMES)
            data_set['year'] = year
        else:
            data = pd.read_csv((URL_PATTERN.format(year)), header=None, names=COLUMN_NAMES)
            data['year'] = year
            data_set = data_set.append(data)
    return data_set
data = load_src()
```
